```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
students = pd.read_csv('./Students.csv')
attendances=pd.read_csv('./Attendances.csv')
subjects=pd.read_csv('./Subjects.csv')
attendances.head()
```

```python
a = pd.merge(attendances, students, how='inner', left_on='student', right_on='id')

a.rename(columns={'id_x':'attendance_id', 'fingerprint_x':'fingerprint','createdAt_x':'date','student':'student_id'},inplace=True)
a.drop(['updatedAt_x','id_y','fingerprint_y','createdAt_y','updatedAt_y'],axis=1,inplace=True)
a.head(1)
```

```python
df = pd.merge(a, subjects, how='inner', left_on='subject', right_on='id')
df.rename(columns={'name_x': 'name', 'semester_x': 'semester', 'branch_x': 'branch', 'name_y': 'subject_name', 'id': 'subject_id'}, inplace=True)
df.drop(['semester_y', 'branch_y', 'createdAt', 'updatedAt', 'subject'], axis=1, inplace=True)
df.head(5)
```

```python
bunty= df.groupby({'fingerprint': 'total_attendance', 'subject_name': 'subject_name'}, axis=1).count()
```
